```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_tmp  = pd.read_excel('Overestimation_scenarios.xlsx')
```

```python
df_tmp
```

```python
np.array(df_tmp['AEI_0 (H2S)'])
```

```python
x_H2S = np.array([1,3,5,7])
y_AEI_H2S = np.array(df_tmp['AEI_0 (H2S)']) 
y_CDO_H2S = np.array(df_tmp['CDO_1 (H2S)'])


plt.figure(dpi = 90)
plt.bar(x_H2S-0.4, y_AEI_H2S*100, width = 0.7, color = [0,0.8, 0.5],
       label = 'AEI-0')
plt.bar(x_H2S+0.4, y_CDO_H2S*100, width = 0.7, color = [0,0.5, 0.8],
       label = 'CDO-1')

plt.ylabel('CH$_4$ recovery (%)', fontsize = 15)
plt.ylim([40,100])
x_scenarios = ['Accurate \n prediction \n scenario',
               '50% \n overestimated \n CH$_4$ scenario',
              '100% \n overestimated \n CH$_4$ scenario',
              '200% \n overestimated \n CH$_4$ scenario']
plt.xticks([1-0.2,3-0.2,5-0.1,7+0.1], x_scenarios, fontsize = 11)
plt.yticks(fontsize = 14)
plt.legend(fontsize = 15)
plt.title('H$_2$S/CH$_4$ Separation Case', fontsize = 15)
```

```python
x_H2S = np.array([1,3,5,7])
y_AEI_H2S = np.array(df_tmp['AEI_0 (CO2)']) 
y_CDO_H2S = np.array(df_tmp['CDO_1 (CO2)'])


plt.figure(dpi = 90)
plt.bar(x_H2S-0.4, y_AEI_H2S*100, width = 0.7, color = [0,0.8, 0.5],
       label = 'AEI-0')
plt.bar(x_H2S+0.4, y_CDO_H2S*100, width = 0.7, color = [0,0.5, 0.8],
       label = 'CDO-1')

plt.ylabel('CH$_4$ recovery (%)', fontsize = 15)
plt.ylim([40,100])
x_scenarios = ['Accurate \n prediction \n scenario',
               '50% \n overestimated \n CH$_4$ scenario',
              '100% \n overestimated \n CH$_4$ scenario',
              '200% \n overestimated \n CH$_4$ scenario']
plt.xticks([1-0.2,3-0.2,5-0.1,7+0.1], x_scenarios, fontsize = 11)
plt.yticks(fontsize = 14)
plt.legend(fontsize = 15)
plt.title('CO$_2$/CH$_4$ Separation Case', fontsize = 15)
```
